```python
import numpy as np
import matplotlib.pyplot as plt
from Biseccion import biseccion
from NewtonRaphson import newton_raphson
from Secante import secante

```

```python
# función y derivada
def g(x):
    return x**2 + 1/(x - 7)

def dg(x):
    return 2*x - 1/((x - 7)**2)
```

```python
# Método de Bisección
# Intervalo [-2, 1]
x_zero, fx, z_list, iters, conv = biseccion(f=g, a=-2, b=0, tol=1e-7, maxIter=100, prt=1)
print(f"Cero aproximado: {x_zero}, f(x): {fx}, iteraciones: {iters}")
```

```python
# Método de la Secante
# Dos puntos iniciales cercanos al cero, por ejemplo -1 y 1
x_zero, fx, z_list, errx, erry, iters, conv = secante(g, -1, 1, tol=1e-7, maxIter=100, prt=1)
print(f"Cero aproximado: {x_zero}, f(x): {fx}, iteraciones: {iters}")
```

```python
# Método de Newton-Raphson
# Punto inicial cercano al cero, por ejemplo x0 = 0.5
x_zero, fx, z_list, errx, erry, iters, conv = newton_raphson(g, dg, 0.5, tol=1e-7, maxIter=100, prt=1)
print(f"Cero aproximado: {x_zero}, f(x): {fx}, iteraciones: {iters}")
```
